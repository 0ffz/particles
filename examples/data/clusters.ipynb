```kotlin
%use dataframe
%use kandy
```

```kotlin
val cluster1 = DataFrame.readCSV("clusters_1.5.csv").add("Temperature") { "Low (1.5)" }
val cluster2 = DataFrame.readCSV("clusters_2.5.csv").add("Temperature") { "Medium (2.5)" }
val cluster3 = DataFrame.readCSV("clusters_3.5.csv").add("Temperature") { "High (3.5)" }
val df = cluster1.concat(cluster2, cluster3).filter { "size"<Int>() < 100 }
df
```

```kotlin
import org.jetbrains.kotlinx.kandy.ir.scale.PositionalScale
import org.jetbrains.kotlinx.kandy.letsplot.layers.builders.subcontext.BorderLine
import org.jetbrains.letsPlot.core.spec.back.transform.bistro.util.scale
import org.jetbrains.letsPlot.themes.themeLight
plot(df.groupBy("Temperature")) {

    statBin("size", binsOption = BinsOption.byNumber(17)) { //, adjust = 5.0
//        fillColor("group") {
//            scale = continuous(Color.BLUE..Color.RED)
//        }
        bars {
            fillColor("Temperature")
//            borderLine {
//                color("group")
//            }
//            alpha = 0.03
            x(Stat.x)
            y(Stat.count.map { sqrt(it.toDouble()) }) //.mapIndexed { index, it -> it * Stat.x.get(index) })
            tooltips(Stat.count)
            themeLight()
        }
        layout {
            title = "Size distribution under increasing temperature"
            yAxisLabel = "count (sqrt)"
            xAxisLabel = "Cluster size (# of monomers)"
        }
        layout.theme = Theme.HIGH_CONTRAST_LIGHT
    }
}
```
